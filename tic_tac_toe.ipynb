```python
# Tic tac toe game
from random import randint

def display_board(board):
    print('  '+board[7]+'  |  '+ board[8]+ '  |  ' + board[9])
    print('----------------')
    print('  '+board[4]+'  |  '+ board[5]+ '  |  ' + board[6])
    print('----------------')
    print('  '+board[1]+'  |  '+ board[2]+ '  |  ' + board[3])

def marker():
    marker = ''
    while not (marker == 'X' or marker == 'O'):
        marker = input("Player: Do you want to be 'X' or 'O'?")
    if marker == 'X':
        return ('X' , 'O')
    else:
        return ('O' , 'X')

def marker_placing(board, marker, position):
    board[position] = marker 

def who_go_first():
    x = randint(0,1)
    if x ==0:
        return 'Player' 
    else:
        return 'Computer'

def space_check(board, position):
    return board[position] == ' '

def full_board_check(board):
    for i in range(1,10):
        if space_check(board, i):
            return False 
    return True

def win_check(board, mark):
    return ((board[7] == mark and board[8] == mark and board[9] == mark) or
            (board[4] == mark and board[5] == mark and board[6] == mark) or
            (board[1] == mark and board[2] == mark and board[3] == mark) or
            (board[1] == mark and board[4] == mark and board[7] == mark) or
            (board[2] == mark and board[5] == mark and board[8] == mark) or
            (board[3] == mark and board[6] == mark and board[9] == mark) or
            (board[1] == mark and board[5] == mark and board[9] == mark) or
            (board[3] == mark and board[5] == mark and board[7] == mark))

def player_choice(board):
    position = 0
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board, position):
        position = int(input("Choose your next position: [1 - 9]:"))
    return position
def computer_play(board):
    position = 0
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board, position):
        position = randint(1,9)
    return position
def replay():
    return input('Do you want to play again? Enter Yes or No: ').lower().startswith('y')

# play the game
print('Welcome to Tic Tac Toe!')
while True:
    the_board = [' ']*10
    player_marker, computer_marker = marker()
    turn = who_go_first()
    print(turn + ' will go first.')
    play_game = input("Are you ready? Enter Yes or No: ")
    if play_game[0].lower() == 'y':
        game_on = True
    else:
        game_on = False 
    while game_on:
        if turn == "Player":
            display_board(the_board)
            position = player_choice(the_board)
            marker_placing(the_board, player_marker, position)
            if win_check(the_board, player_marker):
                display_board(the_board)
                print("Congratulations! You have won the game!")
                game_on = False 
            else: 
                if full_board_check(the_board):
                    print("Draw")
                    break
                else:
                    turn = 'Computer'
        if turn == "Computer":
            display_board(the_board)
            position = computer_play(the_board)
            marker_placing(the_board, computer_marker, position)
            if win_check(the_board, computer_marker):
                display_board(the_board)
                print("Computer have won the game!")
                game_on = False 
            else:
                if full_board_check(the_board):
                    print("Draw")
                    break
                else:
                    turn = 'Player'
    if not replay():
        break

# test_board = [' ']*10
# x = player_choice(test_board)
# marker_placing(test_board, 'X', x)
# display_board(test_board)
```
